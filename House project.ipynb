# Univariate Analysis

```python
def histogram(data, feature):
    sns.distplot(data[feature], kde=False)
```

```python
for i in train_num.columns:
    histogram(train_num, i)
    plt.show()
```

        By seeing the histogram plot for each numerical feature, we can conclude that some of the features are normally distrubuted and some of the features are ordinal (order in ranking example number of cars, number of rooms). The normally distributed features have skewness. So we need to fill the null values with median value.

```python
def num_fill(data):
    num_column = list(data.columns)
    for i in num_column:
        data[i].fillna(data[i].median(), inplace=True)
    return data
```

```python
train_fill_num = num_fill(train_num)
train_fill_num.head()
```

# Outlier Analysis using boxplot

```python
def boxplot(data, feature):
    sns.boxplot(data[feature])
```

```python
for i in train_fill_num.columns:
    boxplot(train_fill_num, i)
    plt.show()
```

        From the boxplot we can conclude that each feature consist of outliers, here most of the outliers are treatable i.e., by error. By interpretting the box plot, we can the skewness of the feature, median, inter quartile range (25% to 50%), outliers can be found and cured.

```python
def cure_outliers(data):
    for i in data.columns:
        q1 = np.nanquantile(data[i], 0.25)
        q3 = np.nanquantile(data[i], 0.75)
        iqr = q3 - q1
        lower_bound = q1 - (1.5 * iqr)
        upper_bound = q3 + (1.5 * iqr)
        lower_quantile = np.nanquantile(data[i], 0.10)
        upper_quantile = np.nanquantile(data[i], 0.90)
        for j in data[i]:
            if j <= lower_bound:
                data[i].replace(to_replace = j, value = lower_quantile, inplace=True)
            elif j >= upper_bound:
                data[i].replace(to_replace = j, value = upper_quantile, inplace=True)
            else:
                data[i].replace(to_replace = j, value = j, inplace=True)
    return data
```

```python
train_cure_out = cure_outliers(train_fill_num)
train_cure_out.head()
```

```python
for i in train_cure_out.columns:
    boxplot(train_cure_out, i)
    plt.show()
```

# One hot encoding and ordinal  encoding

```python
ordinal_columns = []
for i in train_fill_cat.columns:
    if i.endswith('Qual') or i.endswith('Cond') or i.endswith('QC') or i.endswith('Qu'):
        ordinal_columns.append(i)
ordinal_columns
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ordinal_encode = OrdinalEncoder()
ordinal_encoded = ordinal_encode.fit(train_fill_cat[ordinal_columns])
cat_ordinal_encode = ordinal_encoded.transform(train_fill_cat[ordinal_columns])
cat_ordinal_encode
```

```python
cat_encode_columns = pd.DataFrame(cat_ordinal_encode, columns=ordinal_columns)
cat_encode_columns.head()
```

```python
one_hot_columns = []
for i in train_fill_cat.columns:
    if i not in ordinal_columns:
        one_hot_columns.append(i)
one_hot_columns
```

```python
def drop_ordinal(data, feature):
    data.drop(feature, inplace=True, axis=1)
    return data
```

```python
train_one_hot = drop_ordinal(train_fill_cat, ordinal_columns)
train_one_hot.head()
```

```python
cat_one_hot_encode = pd.get_dummies(train_one_hot, columns=one_hot_columns, prefix=one_hot_columns, prefix_sep='_', drop_first=True)
cat_one_hot_encode.head()
```

```python
def join_two_cat(data1, data2):
    data = data1.join(data2)
    return data
```

```python
categorical_encoded_data = join_two_cat(cat_encode_columns, cat_one_hot_encode)
categorical_encoded_data.head()
```

```python
def join_cat_num(data1, data2):
    data = data1.join(data2)
    return data
```

```python
X = join_cat_num(train_cure_out, categorical_encoded_data)
X.head()
```

```python
from scipy.stats import spearmanr
```

```python
cat_corr = []
cat_pval = []
categorical_correlated_columns = []
for i in categorical_encoded_data.columns:
    corr, p_val = spearmanr(categorical_encoded_data[i], y)
    cat_corr.append(corr)
    cat_pval.append(p_val)
for i, j, k in zip(cat_corr, categorical_encoded_data.columns, cat_pval):
    if abs(i) >= 0.5 and abs(k) <= 0.05:
        categorical_correlated_columns.append(j)
categorical_correlated_columns
```

```python
area = []
for i in X.columns:
    if i.endswith('Area'):
        area.append(i)
area
```

```python
def new_features(data):
    data['TotalArea'] = data['LotArea'] + data['MasVnrArea'] + data['GrLivArea'] + data['GarageArea'] + data['PoolArea']
    data['TotalYearSold'] = data['YrSold'] - data['YearBuilt']
    data['ModifyYears'] = data['YrSold'] - data['YearRemodAdd']
    return data
```

# Stacking Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import PassiveAggressiveRegressor
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score
```

```python
lr = LinearRegression()
par = PassiveAggressiveRegressor()
las = Lasso()
rr = RandomForestRegressor()
gbr = GradientBoostingRegressor()
etr = ExtraTreesRegressor()
knr = KNeighborsRegressor()
```

```python
reg = [rr, par, las, gbr, etr, knr]
for algo in reg:
    score = cross_val_score(algo, X_train_scale, y_train, cv=5)
    print("The score for {} is {}.".format(algo, score.mean()))
```

```python
from sklearn.ensemble import StackingRegressor
```

```python
regressor = [('las', las), ('lr', lr), ('gbr', gbr), ('etr', etr), ('knr', knr)]
```

```python
stack_model = StackingRegressor(estimators=regressor, final_estimator=rr)
```

```python
score = cross_val_score(stack_model, X_train_scale, y_train, cv=3, scoring='neg_root_mean_squared_error')
print("The score for {} is {}.".format('stack_model', score.mean()))
```

```python
stack_model.fit(X_train_scale, y_train)
```

```python
stack_model.score(X_train_scale, y_train)
```

```python
stack_model.score(X_test_scale, y_test)
```

```python
stack_pred = stack_model.predict(test_transform)
stack_pred
```

```python
stack_out = pd.DataFrame({'Id':test_id,
                         'SalePrice':stack_pred})
stack_out.head()
```
